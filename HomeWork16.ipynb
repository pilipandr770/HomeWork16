```python
%%writefile train_models.py

import numpy as np
from keras.datasets import imdb
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN, LSTM, Bidirectional, Dense
from keras.callbacks import EarlyStopping
import pickle

# Параметри
max_features = 20000  # Кількість слів у словнику
maxlen = 100  # Максимальна довжина рецензій
batch_size = 32
epochs = 10
input_length = 100

# Завантаження даних
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)

# Визначення моделей
def create_rnn_model(input_length, max_features):
    model = Sequential()
    model.add(Embedding(max_features, 128, input_length=input_length))
    model.add(SimpleRNN(128))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def create_lstm_model(input_length, max_features):
    model = Sequential()
    model.add(Embedding(max_features, 128, input_length=input_length))
    model.add(LSTM(128))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def create_bidirectional_lstm_model(input_length, max_features):
    model = Sequential()
    model.add(Embedding(max_features, 128, input_length=input_length))
    model.add(Bidirectional(LSTM(128)))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def create_deep_rnn_model(input_length, max_features):
    model = Sequential()
    model.add(Embedding(max_features, 128, input_length=input_length))
    model.add(SimpleRNN(128, return_sequences=True))
    model.add(SimpleRNN(128))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Створення моделей
rnn_model = create_rnn_model(input_length, max_features)
lstm_model = create_lstm_model(input_length, max_features)
bidirectional_lstm_model = create_bidirectional_lstm_model(input_length, max_features)
deep_rnn_model = create_deep_rnn_model(input_length, max_features)

# Функція ранньої зупинки
early_stopping = EarlyStopping(monitor='val_loss', patience=2)

# Навчання моделей
histories = {}
models = {
    'RNN': rnn_model,
    'LSTM': lstm_model,
    'Bidirectional_LSTM': bidirectional_lstm_model,
    'Deep_RNN': deep_rnn_model
}

for model_name, model in models.items():
    print(f'Training {model_name} model...')
    history = model.fit(x_train, y_train,
                        batch_size=batch_size,
                        epochs=epochs,
                        validation_split=0.2,
                        callbacks=[early_stopping])
    histories[model_name] = history.history
    model.save(f'{model_name}_model.h5')

# Збереження історій навчання
with open('history.pkl', 'wb') as f:
    pickle.dump(histories, f)
```

```python
%%writefile train_models.py

import numpy as np
from keras.datasets import imdb
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN, LSTM, Bidirectional, Dense
from keras.callbacks import EarlyStopping
import pickle

# Параметри
max_features = 20000  # Кількість слів у словнику
maxlen = 100  # Максимальна довжина рецензій
batch_size = 32
epochs = 10
input_length = 100

# Завантаження даних
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)

# Визначення моделей
def create_rnn_model(input_length, max_features):
    model = Sequential()
    model.add(Embedding(max_features, 128, input_length=input_length))
    model.add(SimpleRNN(128))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def create_lstm_model(input_length, max_features):
    model = Sequential()
    model.add(Embedding(max_features, 128, input_length=input_length))
    model.add(LSTM(128))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def create_bidirectional_lstm_model(input_length, max_features):
    model = Sequential()
    model.add(Embedding(max_features, 128, input_length=input_length))
    model.add(Bidirectional(LSTM(128)))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def create_deep_rnn_model(input_length, max_features):
    model = Sequential()
    model.add(Embedding(max_features, 128, input_length=input_length))
    model.add(SimpleRNN(128, return_sequences=True))
    model.add(SimpleRNN(128))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Створення моделей
rnn_model = create_rnn_model(input_length, max_features)
lstm_model = create_lstm_model(input_length, max_features)
bidirectional_lstm_model = create_bidirectional_lstm_model(input_length, max_features)
deep_rnn_model = create_deep_rnn_model(input_length, max_features)

# Функція ранньої зупинки
early_stopping = EarlyStopping(monitor='val_loss', patience=2)

# Навчання моделей
histories = {}
models = {
    'RNN': rnn_model,
    'LSTM': lstm_model,
    'Bidirectional_LSTM': bidirectional_lstm_model,
    'Deep_RNN': deep_rnn_model
}

for model_name, model in models.items():
    print(f'Training {model_name} model...')
    history = model.fit(x_train, y_train,
                        batch_size=batch_size,
                        epochs=epochs,
                        validation_split=0.2,
                        callbacks=[early_stopping])
    histories[model_name] = history.history
    model.save(f'{model_name}_model.h5')

# Збереження історій навчання
with open('history.pkl', 'wb') as f:
    pickle.dump(histories, f)
```

```python
%%writefile app.py

import streamlit as st
from keras.models import load_model
from keras.preprocessing.sequence import pad_sequences
from keras.datasets import imdb
import numpy as np
import matplotlib.pyplot as plt
import pickle

# Завантаження моделей та історій
models = {
    'RNN': load_model('RNN_model.h5'),
    'LSTM': load_model('LSTM_model.h5'),
    'Bidirectional_LSTM': load_model('Bidirectional_LSTM_model.h5'),
    'Deep_RNN': load_model('Deep_RNN_model.h5')
}

with open('history.pkl', 'rb') as f:
    histories = pickle.load(f)

# Параметри
max_features = 20000
maxlen = 100

# Завантаження даних
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)
x_test = pad_sequences(x_test, maxlen=maxlen)

st.title('IMDb Review Classification Web App')

# Вибір моделі
model_choice = st.selectbox("Choose a model", list(models.keys()))
model = models[model_choice]

# Введення тексту для класифікації
review = st.text_area("Enter your review:")

# Попередня обробка тексту
if review:
    tokens = imdb.get_word_index()
    words = review.lower().split()
    review_sequence = [tokens.get(word, 0) for word in words]
    review_sequence = pad_sequences([review_sequence], maxlen=maxlen)

    # Класифікація тексту
    prediction = model.predict(review_sequence)
    sentiment = "Positive" if prediction[0][0] > 0.5 else "Negative"
    st.write(f"Sentiment: {sentiment} (Probability: {prediction[0][0]:.4f})")

# Відображення графіків функції втрат та точності
def plot_metrics(history):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

    # Графік функції втрат
    ax1.plot(history['loss'], label='loss')
    ax1.plot(history['val_loss'], label='val_loss')
    ax1.set_title('Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()

    # Графік точності
    ax2.plot(history['accuracy'], label='accuracy')
    ax2.plot(history['val_accuracy'], label='val_accuracy')
    ax2.set_title('Accuracy')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()

    st.pyplot(fig)

if st.button('Show Training Metrics'):
    plot_metrics(histories[model_choice])
```
